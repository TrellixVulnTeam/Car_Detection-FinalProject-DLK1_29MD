```python
import os
file_paths = []
flist = os.listdir('./yolov5/runs/detect/exp')
for file in flist:
    fpath = os.path.join('./yolov5/runs/detect/exp', file)
    file_paths.append(fpath)
```

```python
import numpy as np
random = np.random.randint(0, len(file_paths)-1)
img_path = file_paths[random]
print(img_path)
```

```python
plt.figure(figsize=(30, 25))
for i in range(4):
    random = np.random.randint(0, len(file_paths) - 1)
    plt.subplot(2, 2, i + 1)
    img_path = file_paths[random]
    img = plt.imread(img_path)
    plt.imshow(img)
    plt.title('Predictions', size=10, color="black") 
    plt.xticks([])
    plt.yticks([])
```
